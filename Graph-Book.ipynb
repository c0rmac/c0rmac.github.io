```python
import graphviz
import os
```

```python
dot = graphviz.Digraph('markov-graph')
dot.graph_attr.update({'rankdir': 'LR'})
dot.format = 'svg'
dot.attr('node', shape='circle')

dot.node('J', 'J')
dot.node('K', 'K')

dot.edge('K', 'K', label="r")
dot.edge('K', 'J', label="s")
dot.edge('J', 'J', label="1")

dot.render(directory='images/posts/coding-challenge-1', view=True)
dot
```

```python
dot = graphviz.Digraph('tree-graph')
# dot.graph_attr.update({'rankdir': 'LR'})
dot.format = 'svg'
dot.attr('node', shape='circle')

dot.node('start', style='invis', shape='point')
dot.node('J1', 'J')
dot.node('K1', 'K')

dot.node('J2', 'J')
dot.node('K2', 'K')

dot.node('J3', 'J')
dot.node('Kn-1', 'K', style="invis")

dot.edge('start', 'K1', label="  r")
dot.edge('start', 'J1', label=" s")

dot.edge('K1', 'K2', label="  r")
dot.edge('K1', 'J2', label=" s")

dot.edge('K2', 'Kn-1', label="  forever", style="dotted")
dot.edge('K2', 'J3', label=" s")

dot.render(directory='images/posts/coding-challenge-1', view=True)
dot
```
